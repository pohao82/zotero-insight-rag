```python
import re
from collections import Counter

file_name = "PYTLRCQ8.md"
# locate the content between the tags
start_tag = "To shed some light"
end_tag = "Co and Fe variants"

marker_cache = path_marker_cache + file_name

with open(marker_cache) as f:
    marker_text = f.read()

## Show contamination before regex cleanup
html_tags = re.findall(r'<[^>]+>', marker_text)
print(f"HTML artifacts found: {len(html_tags)}")
print(Counter(html_tags).most_common(10))

# 1. Show a specific table section side by side
from IPython.display import display, Markdown

display(Markdown('\n**Before removing the HTML artifacts from marker-pdf parsing**\n'))  # rendered as markdown
start = marker_text.find(start_tag)
end = marker_text.find(end_tag)
text_string = marker_text[start:end]
display(text_string)  # rendered as markdown

display(Markdown('\n**After removing the HTML artifacts from marker-pdf parsing**\n'))  # rendered as markdown
text_string = re.sub(r'\s*<sub>\s*(\d+)\s*</sub>\s*', r'\1', text_string)
display(text_string)  # rendered as markdown
```
